```python
#normalise the images to bring them to scale
N_image = images / 255.0  # Divide by maximum pixel value (255) to scale between 0 and 1

# Check for missing values (NaN or None)
Missing_val = np.isnan(N_image).any() or np.any(np.equal(N_image, None))

if Missing_val:
    # Handle missing values if any
    
    N_image = np.nan_to_num(N_image)
    
print("Normalised image:")
print(N_image[10])
```

```python
labels = mnist["label"][0]
label_counts = [0] * 10  # Initialize a list to store counts for each digit

for i in labels:
    label_counts[int(i)] += 1

# Display the distribution of labels
plt.figure(figsize=(8, 5))
plt.bar(range(10), label_counts, color='green')
plt.xlabel('Digit')
plt.ylabel('Occurence')
plt.xticks(range(10))
plt.title('Distribution of Labels (Digits 0-9)')
plt.grid(axis='y')
plt.show()
```
